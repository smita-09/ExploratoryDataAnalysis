I am using a Linear Regression model to train and predict the salary which is vey basic model. The accuracy is measured using R2 Score.

# Import the Libraries

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

# Read the Data

```python
train_df = pd.read_csv("/kaggle/input/salary/Salary.csv")
train_df.head(10)
```

We can see from here that there are just two features in this dataset and both these features appear to be in float so datatype conversion won't be needed.

# Null Value Check

```python
train_df.isnull().sum()
```

```python
train_df.describe()
```

We could see that the mean value of the yearsexperience lie around 6.3 and mean of Salary lies around 83k. Std is 3 years and for salary it is 32k. The min year exp is 1 and max is 13.5, The min of salary that anyone has is 37k and max is 139k and we can there is a lot of deviation in here.

# Visulaization

Let's do the visualizations to have better understanding of the data

```python
fig = plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
sns.boxplot(y = "YearsExperience", data = train_df, palette="Set3")
plt.subplot(1,2,2)
sns.boxplot(y = "Salary", data = train_df, palette = "Set3")
plt.show()
```

These boxplots give us a idea about median and outliers and visually its more clear now what 25% percentile and what 50% perecentile is for this dataset.

```python
fig = plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
sns.histplot(x = "YearsExperience", data = train_df, palette="Set3")
plt.subplot(1, 2,2)
sns.histplot(x = "Salary", data = train_df, palette = "Set3")
plt.show()
```

These histplots give us a idea about that data is not normally distributed as we could see.

### Target 

```python
y = train_df.pop("Salary")
```

### Splitting the Data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(train_df, y, test_size = 0.2, random_state = 0)
```

### Model Training

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

## Model Prediction

```python

plt.figure(figsize = (20,5))
plt.subplot(1,2,1)
plt.scatter(x_train, y_train, color = 'Grey')
plt.plot(x_train, lr.predict(x_train))
plt.xlabel('Years of Experience')
plt.ylabel('Salary')

plt.subplot(1,2,2)
plt.scatter(x_test,y_test,c = 'Grey', )
plt.plot(x_train, lr.predict(x_train))

plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

### Evaluation

```python
from sklearn.metrics import r2_score
r2_score(y_test,lr.predict(x_test))
```

We could see the score and we could see the accuracy. The better the r2 score better the accuracy. Max value of R2 score is 1
